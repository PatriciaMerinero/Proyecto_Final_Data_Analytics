```python
# Unir los datasets por la columna com√∫n
df_merged = pd.merge(df_trans, df_clientes, on='customer_id', how='left')

# Comprobaci√≥n
print("Filas combinadas:", df_merged.shape)
df_merged.head(10)
```

```python
df_merged['customer_id'].isna().sum()
```

```python


# 0) Ra√≠z del proyecto = carpeta padre de NOTEBOOKS
carpeta_proyecto = Path.cwd().parent

# 1) Directorio de salida (en DATA/DATA_OUTPUT arriba)
carpeta_salida = carpeta_proyecto / "data" / "clean"
carpeta_salida.mkdir(parents=True, exist_ok=True)

# 2) Ruta final del CSV
ruta_csv = carpeta_salida / "datos_unidos.csv"

# 3) Guardar 
df_merged.to_csv(ruta_csv, index=False, encoding="utf-8-sig")

print("Archivo guardado en:", ruta_csv.resolve())
```
