# Comprueblo raiz del proyecto y fijo rutas

```python
# === Rutas robustas para todo el proyecto ===
from pathlib import Path
import os

# 1Ô∏è‚É£ Definir ra√≠z del proyecto seg√∫n ubicaci√≥n del notebook
CWD = Path.cwd()
PROJECT_ROOT = CWD.parent if (CWD.name == "NOTEBOOKS") else CWD

# 2Ô∏è‚É£ Directorios principales
DATA_DIR = PROJECT_ROOT / "DATA"
RAW_DIR = DATA_DIR / "DATA_RAW"
OUT_DIR = DATA_DIR / "DATA_OUTPUT"

# 3Ô∏è‚É£ Crear carpeta de salida si no existe
OUT_DIR.mkdir(parents=True, exist_ok=True)

# 4Ô∏è‚É£ Mostrar estructura actual
print("CWD:          ", CWD)
print("PROJECT_ROOT: ", PROJECT_ROOT)
print("RAW_DIR:      ", RAW_DIR)
print("OUT_DIR:      ", OUT_DIR)

# 5Ô∏è‚É£ Comprobaci√≥n de existencia de los CSV iniciales
clientes_path = RAW_DIR / "clientes_sinteticos.csv"
transacciones_path = RAW_DIR / "transacciones_sinteticas.csv"

print("\nArchivos RAW:")
print("- clientes_sinteticos.csv existe? ", clientes_path.exists())
print("- transacciones_sinteticas.csv existe? ", transacciones_path.exists())
```

# CARGO Y GUARDO ARCHIVOS

```python
df_clientes = pd.read_csv(clientes_path)
df_trans    = pd.read_csv(transacciones_path)

print("Clientes shape:", df_clientes.shape)
print("Transacciones shape:", df_trans.shape)
```

# VISUALIZACI√ìN R√ÅPIDA

```python
# Comprobar nombres de columnas y cantidad de cada dataset
print("CLIENTES:")
print("N√∫mero de columnas:", len(df_clientes.columns))
print(df_clientes.columns.tolist())

print("\nTRANSACCIONES:")
print("N√∫mero de columnas:", len(df_trans.columns))
print(df_trans.columns.tolist())
```

```python
pd.set_option('display.width', 2000)  # aumenta el ancho m√°ximo que se muestra
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)

print("CLIENTES")
print(df_clientes.head(10))

print("\nTRANSACCIONES")  # '\n' a√±ade un salto de l√≠nea para separar visualmente las dos salidas
print(df_trans.head(10))

```

# UNO LOS DOS ARCHIVOS
## 1. Compruebo que columna tienen en com√∫n
## 2. Uno los dataset customer_id
## 3. Verifico si todo esta bien
## 4. Guardo el Dataset unido en RAW OUTPUT

```python
print(df_clientes.columns)
print(df_trans.columns)
```

```python
# Unir los datasets por la columna com√∫n
# (ajusta el nombre del campo si en tus CSV se llama distinto)
df_merged = pd.merge(df_trans, df_clientes, on="customer_id", how="left")

# Mostrar verificaci√≥n b√°sica
print(" Uni√≥n completada correctamente")
print("Filas totales:", df_merged.shape[0])
print("Columnas totales:", df_merged.shape[1])
print("\nColumnas disponibles:", list(df_merged.columns[:10]), "...")

# Guardar en la carpeta can√≥nica DATA_OUTPUT
OUT_FILE = OUT_DIR / "datos_unidos.csv"
df_merged.to_csv(OUT_FILE, index=False, encoding="utf-8-sig")

print("\n Archivo unificado guardado en:", OUT_FILE.resolve())
```
