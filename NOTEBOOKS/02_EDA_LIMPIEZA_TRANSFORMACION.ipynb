### Calculo límites con el método del IQR (Interquartile Range)

```python
# Cálculo del rango intercuartílico
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3 - Q1

# Límites para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Detección
outliers = df[(df['amount'] < limite_inferior) | (df['amount'] > limite_superior)]
print("Número de valores extremos:", len(outliers))
print(outliers[['amount']].head(10))
```

### Interpretación de los resultados

El IQR detecta valores que se salen del rango típico de la variable.

Los valores extremos están en torno a 500 €, así que no parece que sean errores de carga ni importes disparatados.

Probablemente son transacciones altas, pero legítimas.

### Instalo diccionario %pip install -q pycountry. e importo (en Instalaciones)

```python
# Extras para códigos no estándar o históricos que suelen aparecer
EXTRAS = {
    'UK': 'United Kingdom',   # a veces usan UK en vez de GB
    'XK': 'Kosovo',           # código no ISO oficial pero muy usado
    'EL': 'Greece',           # variante europea para Grecia
    'AN': 'Netherlands Antilles',  # histórico
}

def code_to_country(code: str):
    code = code.strip().upper()
    if code in EXTRAS:
        return EXTRAS[code]
    c = pycountry.countries.get(alpha_2=code)
    return c.name if c else None

def normalize_region(val):
    if isinstance(val, str) and re.fullmatch(r'[A-Za-z]{2}', val.strip()):
        # Es un código de 2 letras → convertir a país
        name = code_to_country(val)
        return name if name else val.strip().upper()  # si no se encuentra, deja el código para revisarlo
    # No es código (p.ej. “Texas”, “New York”) → limpito y título
    return val.strip().title() if isinstance(val, str) else val

df['region_normalized'] = df['region'].apply(normalize_region)
```

### Comprobación

```python
# Códigos de 2 letras presentes
codes = sorted(set(df.loc[df['region'].str.fullmatch(r'[A-Za-z]{2}', na=False), 'region'].str.upper()))
# Códigos que no se mapearon (si los hay)
unmapped = [c for c in codes if not code_to_country(c)]

print(f"Códigos de 2 letras detectados: {len(codes)} → no mapeados: {len(unmapped)}")
print("No mapeados:", unmapped[:20])  # muestra los primeros si hubiera
print(df[['region','region_normalized']].head(10))
```

## 20. Country
### Ver países únicos

```python
countries = df['country'].unique()
print("Ejemplos de países registrados:")
print(countries[:50])  
```

### Normalizo formato

```python
df['country'] = df['country'].astype(str).str.strip().str.title()
```

### Limpieza y estandarización

```python
def normalize_country(val):
    if not isinstance(val, str) or val.strip() == "":
        return None
    val = val.strip().upper()
    try:
        country = pycountry.countries.get(alpha_2=val)
        if country:
            return country.name
        country = pycountry.countries.lookup(val)
        return country.name
    except:
        # Por si hay abreviaturas especiales o no estándar
        extras = {
            'UK': 'United Kingdom',
            'USA': 'United States',
            'UAE': 'United Arab Emirates',
            'KOR': 'South Korea',
            'RUS': 'Russia',
            'CZE': 'Czech Republic',
            'VN': 'Vietnam',
            'IRL': 'Ireland',
            'KSA': 'Saudi Arabia'
        }
        return extras.get(val, val.title())  # deja el valor original si no lo encuentra

df['country_normalized'] = df['country'].apply(normalize_country)
```

### Compruebo resultados

```python
print("Registros distintos antes:", df['country'].nunique())
print("Registros distintos después:", df['country_normalized'].nunique())

print("\nMuestra comparativa:")
print(df[['country', 'country_normalized']].head(20))
```

### Clasifico en categorías de riesgo (opcional)

```python
def categorize_risk(score):
    if pd.isna(score):
        return "Desconocido"
    elif score < 0.3:
        return "Bajo"
    elif score < 0.7:
        return "Medio"
    else:
        return "Alto"

df['risk_level'] = df['risk_score'].apply(categorize_risk)
```
