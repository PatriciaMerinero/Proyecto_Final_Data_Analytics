## 4. Amount= float
### Miro valores extremos


```python
df['amount'].describe()

```

### Calculo límites con el método del IQR (Interquartile Range)

```python
# Cálculo del rango intercuartílico
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3 - Q1

# Límites para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Detección
outliers = df[(df['amount'] < limite_inferior) | (df['amount'] > limite_superior)]
print("Número de valores extremos:", len(outliers))
print(outliers[['amount']].head(10))
```

### Interpretación de los resultados

El IQR detecta valores que se salen del rango típico de la variable.

Los valores extremos están en torno a 500 €, así que no parece que sean errores de carga ni importes disparatados.

Probablemente son transacciones altas, pero legítimas.

### Visualizo

```python


plt.figure(figsize=(6,4))
plt.boxplot(df['amount'], vert=False)
plt.title("Distribución de importes (amount)")
plt.xlabel("Importe (€)")
plt.show()
```

La mayoría de las operaciones se concentran entre 100 € y 300 €,

la mediana ronda los 200 €,

y existen unas pocas transacciones de 500 € o más que son las que marcan los puntos fuera del rango.
