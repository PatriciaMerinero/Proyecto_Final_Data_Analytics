### Calculo l√≠mites con el m√©todo del IQR (Interquartile Range)

````python
# C√°lculo del rango intercuart√≠lico
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3 - Q1

# L√≠mites para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Detecci√≥n
outliers = df[(df['amount'] < limite_inferior) | (df['amount'] > limite_superior)]
print("N√∫mero de valores extremos:", len(outliers))
print(outliers[['amount']].head(10))
````

### Resultado

El IQR detecta valores que se salen del rango t√≠pico de la variable.

Los valores extremos est√°n en torno a 500 ‚Ç¨, as√≠ que no parece que sean errores de carga ni importes disparatados.

Probablemente son transacciones altas, pero leg√≠timas.

````python
# Extras para c√≥digos no est√°ndar o hist√≥ricos que suelen aparecer
EXTRAS = {
    'UK': 'United Kingdom',   # a veces usan UK en vez de GB
    'XK': 'Kosovo',           # c√≥digo no ISO oficial pero muy usado
    'EL': 'Greece',           # variante europea para Grecia
    'AN': 'Netherlands Antilles',  # hist√≥rico
}

def code_to_country(code: str):
    code = code.strip().upper()
    if code in EXTRAS:
        return EXTRAS[code]
    c = pycountry.countries.get(alpha_2=code)
    return c.name if c else None

def normalize_region(val):
    if isinstance(val, str) and re.fullmatch(r'[A-Za-z]{2}', val.strip()):
        # Es un c√≥digo de 2 letras ‚Üí convertir a pa√≠s
        name = code_to_country(val)
        return name if name else val.strip().upper()  # si no se encuentra, deja el c√≥digo para revisarlo
    # No es c√≥digo (p.ej. ‚ÄúTexas‚Äù, ‚ÄúNew York‚Äù) ‚Üí limpito y t√≠tulo
    return val.strip().title() if isinstance(val, str) else val

df['region_normalized'] = df['region'].apply(normalize_region)
````

### Limpieza y estandarizaci√≥n

````python
def normalize_country(val):
    if not isinstance(val, str) or val.strip() == "":
        return None
    val = val.strip().upper()
    try:
        country = pycountry.countries.get(alpha_2=val)
        if country:
            return country.name
        country = pycountry.countries.lookup(val)
        return country.name
    except:
        # Por si hay abreviaturas especiales o no est√°ndar
        extras = {
            'UK': 'United Kingdom',
            'USA': 'United States',
            'UAE': 'United Arab Emirates',
            'KOR': 'South Korea',
            'RUS': 'Russia',
            'CZE': 'Czech Republic',
            'VN': 'Vietnam',
            'IRL': 'Ireland',
            'KSA': 'Saudi Arabia'
        }
        return extras.get(val, val.title())  # deja el valor original si no lo encuentra

df['country_normalized'] = df['country'].apply(normalize_country)
````
